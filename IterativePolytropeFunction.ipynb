```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import matplotlib.cm as cm
```

```python
from le_fun import getpoly
```

```python
n_list = [1e-10, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
colours = cm.inferno(np.linspace(0, 0.9, len(n_list)))

for i, n in enumerate(n_list):
    r, rho, m, P = getpoly(n)
    theta = rho/rho[0]
    plt.plot(r, theta, label=f"n={n}", color=colours[i])
plt.legend()
plt.show()
```
